```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
file_path = "Marketing_Design_Dataset (1).csv"
df = pd.read_csv(file_path)

# Generate descriptive statistics
descriptive_stats = df.describe(include='all')

# Display the statistics
print(descriptive_stats)

# Create histogram plot
plt.figure(figsize=(10, 6))
sns.histplot(df['Campaign_Budget'], bins=30, kde=True)
plt.xlabel('Campaign Budget')
plt.ylabel('Frequency')
plt.title('Distribution of Campaign Budget')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "Marketing_Design_Dataset (1).csv"  # Update with your actual file path
df = pd.read_csv(file_path)

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df["Social_Media_Followers"], df["Customer_Retention_Rate"], alpha=0.5, color="blue")
plt.xlabel("Social Media Followers")
plt.ylabel("Customer Retention Rate (%)")
plt.title("Scatter Plot of Social Media Followers vs. Customer Retention Rate")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "Marketing_Design_Dataset (1).csv"  # Update with your actual file path
df = pd.read_csv(file_path)

# Create a bar plot with campaign types on the x-axis and count on the y-axis
plt.figure(figsize=(12, 6))
sns.countplot(x='Campaign_Type', data=df)

# Customizing the plot
plt.title("Count of Each Campaign Type")
plt.xlabel("Campaign Types")
plt.ylabel("Count")
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.show()
```
